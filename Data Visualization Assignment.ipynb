```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
#importing the datset
iris = datasets.load_iris()
```

```python
iris_training = iris['data']
iris_output = iris['target']

df_training = pd.DataFrame(iris_training, columns = ['sepal length', 'sepal width', 'petal length', 'petal width'])
```

```python
#reducing the dimensions of the data using principal component analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
df_training = pca.fit_transform(df_training)
```

```python
data = df_training.T
```

```python
x = data[0]
y = data[1]
z = data[2]
```

```python
fig = plt.figure(1, figsize=(4, 3))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=45, azim=134)

plt.cla()

# Reorder the labels to have colors matching the cluster results
ax.scatter(x, y, z, c=iris_output, edgecolor='k')
#plt.show()
```
